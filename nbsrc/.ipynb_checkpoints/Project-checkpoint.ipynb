## Merge Data

```python
df1 = pd.merge(pdVle, pdStudentVle)
df2 = pd.merge(df1, pdCourses)
df3 = pd.merge(df2, pdStudentRegistration)
df4 = pd.merge(df3, pdStudentInfo)
df5 = pd.merge(pdAssessments, pdStudentAssessment)
df6 = pd.merge(df4, df5)
df6
```
